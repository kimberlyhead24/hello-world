```python
# Writing file to test function. Instructions were to hardode, but
# I want to step it up and make it reusable code that we can add five to the grades
# of any file we choose that has a similar format of name then grade on each line
with open('class_scores.txt', 'w') as f:
    print('GWashington 83', file=f)
    print('JAdams 86', file=f)
    print('JMcKinley 85', file=f)
    print('TLincoln 92', file=f)
    print('FHamilton 78', file=f)
    print('EFranklin 90', file=f)
    print('KRoosevelt 87', file=f)
    print('AMadison 95', file=f)
    print('JMonroe 80', file=f)
    print('AMadison 95', file=f)
with open('class_scores.txt', 'r') as f:
    print(f.read())

# Open both files with the correct read or write privileges, then write a list comprehension that takes the input file
# and turns the second part of the line which is grade into an integer and add five
# write this to the new file with the same format as it was in before
def add_five(filename_in, filename_out):
    with open(filename_in, 'r') as file_in, open(filename_out, 'w') as file_out:
        [print(name, int(score)+5, file=file_out) for name, score in [line.strip().split() for line in file_in]]
    with open(filename_out, 'r') as f:
        print(f.read())

# Call the function with the filenames
add_five('class_scores.txt', 'scores2.txt')
```

```python
# Creating a function to find all words that start with gn or kn in a file
def startswith_gn_kn(filename):
  gn_kn_list = [word for line in open(filename, 'r') for word in line.split() if word.startswith('gn') or word.startswith('kn')]
  for word in gn_kn_list:
    print(word)

#call the function for wordlist.txt
startswith_gn_kn('wordlist.txt')
```

```python
# Create a function that determinea what percentage of words start with a vowel
def vowel_percentage(filename):
  vowel_list = [word for line in open(filename, 'r') for word in line.split() if word[0] in 'aeiou']
  vowel_percentage = len(vowel_list)/len(open(filename, 'r').read().split())*100
  print("{:.0f}%".format(vowel_percentage))

#call the function for wordlist.txt
vowel_percentage('wordlist.txt')
```

```python
# Create function to print all 7-letter words that start with th and end in ly.
def seven_letter_words(filename):
  seven_letter_list = [word for line in open(filename, 'r') for word in line.split() if len(word) == 7 and word.startswith('th') and word.endswith('ly')]
  for word in seven_letter_list:
    print(word)

#call the function for wordlist.txt
seven_letter_words('wordlist.txt')
```

```python
# Define function to print the first ten words that start with q
def ten_qwords(filename):
    q_words = [word for line in open(filename, 'r') for word in line.split() if word.startswith('q')]

    for word in q_words[:10]:
        print(word)

ten_qwords('wordlist.txt')
```
