```python
# Create a fuction called rectangle that takes 2 integers m and n
# as arguments and prints out a m*n box consisting of asterisks

def rectangle(m, n):
    for i in range(m):
      print('*' * n)

rectangle(2, 4)
rectangle(5, 10)
```

```python
# Create a function called add_excitement that takes a list of strings
# and adds a ! to the end of each string in the list.
# We will be modifying the original list with no return

def add_excitement(list):
    for i in range(len(list)):
        list[i] = list[i] + '!'
    print(list)

list1 = ['hello', 'world', 'this', 'is', 'a', 'list']
add_excitement(list1)

list2 = ['this', 'is', 'a', 'different', 'list']
add_excitement(list2)
```

```python
# Create a function called add_excitement that takes a list of strings
# and adds a ! to the end of each string in the list.
# We will creating and returning a new list
def add_excitement_newlist(list):
  new_list = []
  for item in list:
        new_list.append(item + '!')
  return new_list


list3 = ['I am very excited', 'I am graduating Python 2', 'Now I am a professional']
new_list1 = add_excitement_newlist(list3)
print(new_list1)

list4 = ['My career is right around the corner', 'It is time to celebrate', 'By coding and creating cool software']
new_list2 = add_excitement_newlist(list4)
print(new_list2)
```

```python
# Create a function that finds the digital root,
# which adds the sum of the digits until only one remains
def digital_root(n):
    while len(str(n)) > 1:
        n = sum(int(digit) for digit in str(n))
    return n

digital_root(15467)

```

```python
# Write a function called firs_diff that is given 2 strings and returns
# the location in which the strings differ, identical would be -1

def first_diff(string1, string2):
    min_len = min(len(string1), len(string2))
    for i in range(min_len):
        if string1[i] != string2[i]:
            return i
    if len(string1) != len(string2):
        return min_len
    return -1
print(first_diff('butter', 'butterfly'))
print(first_diff('butter', 'butterscotch'))
print(first_diff('butter', 'butter'))
print(first_diff('flap', 'flip'))
```
