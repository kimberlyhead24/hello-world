```python
# Product Price Tracker
# Making a dictionary for products and prices that are inputted by the user
Product_info = {}
while True:
    product_list=input("Enter product names and prices (E.g Sally Shampoo, 25.99): ")
    if product_list == 'quit':
        break
    product_name,product_price = product_list.split(',')
    Product_info[product_name] = product_price
print(Product_info)

# Allowing the user to the enter a product and acquire the price from the dictionary
while True:
  product=input("Enter product for price: ")
  if product == 'quit':
    break
  if product in Product_info:
    print(Product_info[product])
  else:
    print("Product not found")
```

```python
# The user wants to find all products with a maximum price that they input
# We will list what is available with the price constraint

price_for_product = input("What is the max price?: ")
for product in Product_info:
  if float(Product_info[product]) <= float(price_for_product):
    print(product)
```

```python
days = {'January':31, 'February':28, 'March':31,
'April':30, 'May':31, 'June':30, 'July':31,
'August':31, 'September':30, 'October':31,
'November':30, 'December':31}

# a) get a month from the user and show them how many days are in the month
while True:
    user_input = input("Enter a month: ")
    if user_input == 'quit':
        break
    else :
      print(days[user_input])
print()

# b) print out all keys in alphabetical order
sorted_days = sorted(days)
print("Months in Alphabteical Order;")
for month in sorted(days):
    print(month)
print()

# c) Print out all months that are 31 days
print("Months with 31 days:")
for month in days:
    if days[month] == 31:
        print(month)
print()

def by_days(item):
  return item[1]
# d) Print out the key and values sorted by number of days
print("Months sorted by number of days: ")
for month, num_days in sorted(days.items(), key=by_days, reverse=True):
    print(month, num_days)

# e) Modify the program from part a) and the dictionary so that the user
# does not have to know how to spell the month name exactly. That is, all
# they have to do is spell the first three letters of the month correctly.
while True:
    user_input = input("Enter a month: ")
    if user_input == 'quit':
        break
    user_prefix = user_input[:3]     # get just first 3 letters
    found = False
    for month in days:
      if month[:3] == user_prefix:   # match the first 3 letters to the month
        print(f"{month} has {days[month]} days.")
        found = True
        break
    if not found:
      print("Month not recognized, try again.")
print()




```
